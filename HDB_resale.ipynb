## HDB Flat Resale Price 2017- February 2025

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hdb = pd.read_csv('resale_hdb_2017_onwards.csv')
hdb[['year', 'month']] = hdb['month'].str.split('-', expand=True)
```

### Selected Towns Line Graph Analysis

```python
selected_towns = ['BEDOK', 'JURONG WEST', 'WOODLANDS','PUNGGOL']

filtered_hdb = hdb[hdb['town'].isin(selected_towns)]

filtered_hdb.head()
```

```python
sns.relplot(x=filtered_hdb['year'], y=filtered_hdb['resale_price'],
            marker='',
            data=filtered_hdb,
            hue='town',
            row='flat_type',
            kind='line')
```

### Line Graph for all Towns

```python
sns.relplot(x=hdb['year'], y=hdb['resale_price'],
            marker='',
            data=hdb,
            hue='town',
            row='flat_type',
            kind='line')
```
